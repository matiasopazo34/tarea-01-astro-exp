```python
class Numbers(object):
    """
    Class containing two floats that computes their product & sum.

    Args:
        a (float): number 1
        b (float): number 2
    """

    def __init__(self, a, b):
        self.a = a
        self.b = b

    def __repr__(self):
        return "The numbers {} and {}.".format(self.a, self.b)

    def num_sum(self):
        """
        Compute the sum of `a` and `b`

        Returns:
            float: sum of `a` and `b`
        """
        return self.a + self.b

    def num_prod(self):
        """
        Compute the product of `a` and `b`

        Returns:
            float: product of `a` and `b`
        """
        return self.a * self.b
```

# 3. Ejemplos de Fiteos

Ajustar una curva a una serie de puntos es una de las formas mas utiles para, desde el conocimento acerca de la teoría, obtener los paramentros fisicos asociados a los procesos que rigen el origen de nuestras observaciones.

En este código encontraras 3 herramientas de scipy para hacerlo.

## Comportamiento teórico real

En el entendimiento de la fisica y la astronomía se utilizan muchas fórmulas para modelar datos. Ahora veremos cómo una curva, según la forma que sepamos de antemano, se ajusta mejor a un set de datos.

Para esto definiremos una recta "ideal". Esta define exactamente como se comportan y relacionan 2 variables: una fuente emisora por ejemplo podría aproximarse, dentro de cierto rango, como un flujo espectral lineal (o cualquier polinomio) en frecuencia de la forma $F_{\nu} = A \nu+B$.

Esto cumple que el flujo energético en erg/s/cm$^2$, por ejemplo, entre las frecuencias $\nu_1$ y $\nu_2$ está dado por

$\int_{\nu_1}^{\nu_2} F_{\nu}\ d\nu\ =\ \int_{\nu_1}^{\nu_2}(A \nu+B) d\nu$

Es algo así como una densidad de flujo en frecuencia.

## a) Lineal

Definimos 50 puntos que "sigan" la forma de la curva, pero añadiendo el factor aleatorio, tal como lo es el ruido, por ejemplo.

```python
np.random.seed(179)

m_true = 1.89  # Pendiente real
n_true = 2.556  # Coeficiente numérico real
f_true = 0.453  # Factor para alejarnos de la curva, representa en el fondo la incerteza de la medición de CADA punto

N = 20  # Numero de puntos aleatorios considerados como "muestra aleatoria" de la variable x

x = np.sort(10*np.random.rand(N))  # Eje x aleatorio, usando np.random para obtener un array de N valores
                                   # entre [0,1), multiplicados por 10 para extender a [0, 10) y usando
                                   # np.sort para ordenarlos de forma ascendente. x es también un array.

y = m_true * x + n_true  # Definimos nuestro valor real de y en torno a los parametros reales y sus valores x.

# Usualmente el error es proporcional al flujo, mientras más brillante más error hay.

dy = f_true * np.random.randn(N)  # Al multiplicar por np.random, alejamos cada punto de la recta por
                                              # un factor cualquiera en el intervalo [0, f_true*|y|) de manera aleatoria.


y += dy  # sumamos error aleatorio. Este nuevo "y" será nuestro valor "medido" con un error asociado
```

```python
x_toplot = np.linspace(0, 10, 50)  # Creamos nuestro eje x desde [0,10] con 500 valores intermedios

plt.figure(figsize=(10,10))  # Graficamos los puntos para ver su distribución
plt.plot(x, m_true*x+n_true,'k')  # Graficamos la recta "real" si tuviesemos precisión infinita (o bien miramos suficiente tiempo para reducir errores aleatorios)
plt.plot(x, y, 'k.')
plt.show()
```

**Métricas de los fiteos a parámetros según un modelo:**

---



Cuando queremos ajustar una curva, esta puede optar por muchos valores de los parámetros, algunos que predicen mejor los datos y otros que no. Para cuantificar el "qué tan bien predice los datos" uno utiliza métricas del ajuste.

La primera métrica útil a definir es la métrica chi_cuadrado. Para esto creamos una función "chi cuadrado" que busca obtener un número (métrica del ajuste) que indica qué tan lejos están los datos medidos (muestra aleatoria) de sus valores teóricos predichos según la ecuación del modelo con parámetros "params" (en este caso recta: m, n pero pueden definir su modelo como quieran, los parámetros del modelo pueden ser hasta límites de una integral).

Lógicamente un algoritmo de "fiteo" que opere con la métrica chi cuadrado probará parámetros del modelo hasta minimizar el chi cuadrado.

A los diferentes algoritmos uno puede darle métricas distintas que el $\chi ^2$ para minimizar (no a todos).

```python
def chi_cuadrado(params):
    m, n = params  # Definimos los parametros de la recta: pendiente, coeficiente numérico
                   # a partir del argumento

    y_model = m * x + n  # Definimos la forma de nuestro modelo: una recta unidimensional

    y_data = y  # Definimos cual es nuestro "y" medido, el cual posee un error aleatorio

    a = 0  # Usamos un contador para ir sumando el chi_cuadrado de cada punto
    for i in range(len(y_data)):  # Iteramos sobre todos los puntos
        a += (y_model[i] - y[i])**2  # La distancia de cada observación al modelo se eleva al cuadrado para tener
                                     # solo valores positivos

    return a  # La funcion retorna esta suma de desviaciones del modelo al cuadrado
```

#### Algoritmo "fmin"

Primero utilizaremos la función fmin (pueden ver como procede usando help(fmin) o en su pagina correspondiente), la cual buscará minimizar esta función y entregarnos los parametros que marquen el chi cuadrado mas pequeño. Para esto, la función necesita un punto de partida, que para nuestro caso usaremos los parametros reales desviados un valor aleatorio entre [0 ,0.1 [.

```python
initial = np.array([m_true, n_true]) + 0.1*np.random.randn(2) # Definimos nuestros parametros iniciales
optimo = fmin(chi_cuadrado, initial)                          # Guardamos estos parametros en la variable optimo
print(optimo)
```

Luego de iterar varias veces compara cuales funciones tienen el menor chi cuadrado y entrega el parametro buscado. Definimos desde "optimo" los valores para la recta y graficamos.

```python
m_est = optimo[0]  # Pendiente obtenida del fiteo
n_est = optimo[1]  # Coef. numerico obtenido del fiteo

plt.figure(figsize=(10,10))
plt.plot(x, y, 'k.')  # Gráfico de los puntos aleatorios (puntos negros)
plt.plot(x, m_true*x+n_true, 'k-', label="real")  # Gráfico de la curva real
plt.plot(x, m_est*x+n_est, 'g-', label="fmin fitted")  # Gráfico de la curva fiteada a los puntos.
plt.legend()
plt.show()
```

Como podemos ver, la curva ajustada no es necesariamente tan parecida a la curva real (relación intrínseca). Nuestro limitante es la cantidad de puntos y la resolucion (desviacion de la recta real) que presentan nuestras observaciones.

Mas puntos/tiempo = Mejor calidad

#### Algoritmo "minimize"

Procedemos con la función minimize, con el mismo objetivo y el mismo procedimiento a la hora de definir nustra función y parametros iniciales.

```python
def chi_minimize(params, x, y):
    m, n = params
    y_model = m*x + n
    y_data = y
    a = 0
    for i in range(len(y_data)):
        a += (y_model[i] - y[i])**2
    return a
```

Con la diferencia que a esta función le decimos que parametros dentro del chi cuadrado son valores variables, y no parte del modelo. Entregandonos aun mas información que fmin.

```python
initial = np.array([m_true, n_true]) + 0.1*np.random.randn(2)
optimo2 = minimize(chi_minimize, initial, args=(x, y))
print(optimo2)
```

Extraemos los parámetros

```python
parametros = optimo2.x
```

Graficamos de la misma forma

```python
plt.figure(figsize=(10,10))

plt.plot(x, y, 'k.')
plt.plot(x, m_true*x+n_true, 'k-', label="real")
plt.plot(x, m_est*x+n_est, 'g-', label='fmin fitted')
plt.plot(x, parametros[0]*x+parametros[1], 'r-', label="minimize fitted")
plt.legend()
plt.show()
```

#### Algoritmo "curve_fit"
Ahora usaremos curve fit, un poco diferente. Se tiene que definir la función y se buscan los parametros que tengan un menor error.

```python
def recta(x,m,n):
    y = m*x + n
    return y

curve_param, covarianza = curve_fit(recta, x, y)
print(curve_param,covarianza)
```

Para hacer el ajuste gaussiano, definimos la función gaussiana al igual que con la recta y colocamos como parámetros todas las variables que no sean la primera

```python
# Se define funcion gaussiana a fitear
def f_gauss(x,T0,mean,stdv):
    return T0*np.exp(-((x-mean)**2)/(2*(stdv**2)))
```

```python
# First Guess: T0=20, m=10, s=1 obtenidos al inspeccionar espectro ploteado
fg = [20, 10, 1]

coefs,cov = curve_fit(f_gauss,v,T, p0=fg) # Se fitea

t0,M,S = coefs[0],coefs[1],coefs[2]  # Se extraen los coeficientes fiteados

print ('Valores fiteados: (t0,M,S) =',(t0,M,S))
```

```python
plt.plot(v,T, label='Data')
plt.plot(v,f_gauss(v,t0, M,S), label='Fiteo')
plt.title('Espectro', fontsize=18)
plt.ylabel('Temperatura [K]', fontsize=18)
plt.xlabel(r'Velocidad [$\frac{km}{s}$]', fontsize=18)
plt.legend()
plt.show()
```

Esta gaussiana es de la forma $T(v)\ =\ T_0\ exp\left(-\frac{(V-M)^2}{2S^2}\right)$ cuya integral (área) es simplemente $\sqrt{2\pi}\ \times\ T_0 \times S$ pero si lo desean pueden buscar la función trapz para integrar manualmente el área bajo la curva fiteada entre rangos razonables (como es gaussiana, desde el promedio hasta $\pm$ 5 $\sigma$ ya abarca casi toda la curva, donde $\sigma=S$ )

### Como leer un cubo de datos


```python
# Solo para omitir warnings, en general NO deberían usarlo
import warnings
warnings.filterwarnings("ignore")

from astropy.io import fits
```

Esta es una forma "automatizada" de obtener la velocidad(v) longitud(l) y latitud(b)

Esta funcion les entrega un arreglo con los valores reales del eje correspondiente


1 es velocidad, 2 es longitud galactica, 3 es latitud galactica

```python
def values(h,j):
    N=h['NAXIS'+str(j)];
    val=np.zeros(N)
    for i in range(0,N):
        val[i] = (i+1-float(h['CRPIX'+str(j)]))*float(h['CDELT'+str(j)]) + float(h['CRVAL'+str(j)])
    return val
```

```python
cubo = fits.open(root + "astroexp/southgal_fixbadc.fits")  # abrir objeto cubo de datos
data = cubo[0].data  # extraer matriz de datos
header= cubo[0].header # extraer el header del archivo fits
```

Estos seran los tres arreglos con los valores reales de los tres ejes del cubo



```python
velocidad=values(header,1)
longitud=values(header,2)
latitud=values(header,3)

#print ('v =',velocidad)
#print ('l =',longitud)
#print ('b =',latitud)

#esta es una forma manual de obtener v, l, b, leyendo el header
#velocidad= np.linspace(-230.7985, 165.8235, 306)
#longitud= np.linspace(300, 348., 385)

```
